### Websocket Live-stream

```python
# Cell 9: WebSocket Listener for Real-Time Market Data (Direct Execution)
import json
import asyncio
import websockets
import datetime
import sqlite3
import time
import os

async def listen_to_market_websocket(duration_minutes=3):
    """Listen to WebSocket for real-time market data"""

    ws_url = os.getenv("WS_CLOB_URL", "wss://ws-subscriptions-clob.polymarket.com/ws/market")
    db_path = os.getenv("DB_PATH", "/content/polymarket_whales.db")

    # Get first active market with clob_token_ids
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    cursor.execute("""
    SELECT id, clob_token_ids, question
    FROM markets
    WHERE active = 1 AND clob_token_ids IS NOT NULL
    LIMIT 1
    """)

    market_data = cursor.fetchone()

    if not market_data:
        print("❌ No active markets found with CLOB token IDs")
        return

    market_id, clob_token_ids_str, question = market_data

    # Parse CLOB token IDs
    try:
        clob_token_ids = json.loads(clob_token_ids_str)
    except:
        print(f"❌ Failed to parse CLOB token IDs: {clob_token_ids_str}")
        return

    print(f"📡 Starting WebSocket listener for market: {question[:60]}...")
    print(f"   Market ID: {market_id}")
    print(f"   Duration: {duration_minutes} minutes")
    print("="*60)

    start_time = datetime.datetime.now()
    end_time = start_time + datetime.timedelta(minutes=duration_minutes)
    last_ping = datetime.datetime.now()
    messages_received = 0

    try:
        async with websockets.connect(ws_url) as websocket:
            # Subscribe to market
            await websocket.send(json.dumps({
                "assets_ids": clob_token_ids,
                "type": "market"
            }))

            print("✅ Connected and subscribed to market feed")
            print("📊 Live feed started...")
            print("-"*60)

            while datetime.datetime.now() < end_time:
                try:
                    # Set timeout to handle ping/pong
                    message = await asyncio.wait_for(websocket.recv(), timeout=1.0)

                    # Check if it's a PONG message
                    if message == "PONG":
                        last_ping = datetime.datetime.now()
                        continue

                    try:
                        data = json.loads(message)
                        messages_received += 1

                        # Handle different message formats
                        if isinstance(data, list):
                            # If it's a list, process each item
                            for item in data:
                                if isinstance(item, dict):
                                    # Store in database
                                    cursor.execute("""
                                    INSERT INTO ws_messages (market_id, asset_id, message_type, data, timestamp)
                                    VALUES (?, ?, ?, ?, ?)
                                    """, (
                                        market_id,
                                        item.get('asset_id', ''),
                                        item.get('event_type', 'market_snapshot'),
                                        json.dumps(item),
                                        item.get('timestamp', int(time.time() * 1000))
                                    ))

                                    # Print live update
                                    if 'bids' in item and 'asks' in item:
                                        best_bid = item['bids'][0]['price'] if item['bids'] else 'N/A'
                                        best_ask = item['asks'][0]['price'] if item['asks'] else 'N/A'
                                        print(f"📈 SNAPSHOT | Bid: {best_bid} | Ask: {best_ask}")

                        elif isinstance(data, dict):
                            # Store in database
                            cursor.execute("""
                            INSERT INTO ws_messages (market_id, asset_id, message_type, data, timestamp)
                            VALUES (?, ?, ?, ?, ?)
                            """, (
                                market_id,
                                data.get('asset_id', ''),
                                data.get('event_type', 'unknown'),
                                json.dumps(data),
                                data.get('timestamp', int(time.time() * 1000))
                            ))

                            # Print live updates based on event type
                            event_type = data.get('event_type', '')

                            if event_type == 'price_change' and 'price_changes' in data:
                                for change in data['price_changes']:
                                    price = change.get('price', 'N/A')
                                    size = change.get('size', 'N/A')
                                    side = change.get('side', 'N/A')
                                    print(f"💹 TRADE | {side} | Price: {price} | Size: {size}")

                            elif 'market' in data and 'bids' in data:
                                best_bid = data['bids'][0]['price'] if data.get('bids') else 'N/A'
                                best_ask = data['asks'][0]['price'] if data.get('asks') else 'N/A'
                                print(f"📊 UPDATE | Bid: {best_bid} | Ask: {best_ask}")

                            else:
                                # Generic message
                                print(f"📨 MSG #{messages_received} | Type: {event_type}")

                        # Commit every 50 messages
                        if messages_received % 50 == 0:
                            conn.commit()
                            elapsed = (datetime.datetime.now() - start_time).total_seconds()
                            remaining = (end_time - datetime.datetime.now()).total_seconds()
                            print(f"⏱️ Status: {messages_received} msgs | {elapsed:.0f}s elapsed | {remaining:.0f}s remaining")

                    except json.JSONDecodeError:
                        print(f"⚠️ Failed to parse message")
                    except Exception as e:
                        print(f"⚠️ Error processing message: {e}")

                    # Send ping if needed
                    if datetime.datetime.now() - last_ping > datetime.timedelta(seconds=10):
                        await websocket.send("PING")
                        last_ping = datetime.datetime.now()

                except asyncio.TimeoutError:
                    # Check if we should continue
                    if datetime.datetime.now() >= end_time:
                        break

                    # Send ping to keep connection alive
                    if datetime.datetime.now() - last_ping > datetime.timedelta(seconds=10):
                        await websocket.send("PING")
                        last_ping = datetime.datetime.now()

                except Exception as e:
                    print(f"⚠️ Error in message loop: {e}")

    except Exception as e:
        print(f"❌ WebSocket connection error: {e}")

    finally:
        conn.commit()
        conn.close()

        elapsed = (datetime.datetime.now() - start_time).total_seconds()
        print("-"*60)
        print(f"✅ WebSocket listener stopped")
        print(f"📊 Final Statistics:")
        print(f"   Total messages: {messages_received}")
        print(f"   Duration: {elapsed:.1f} seconds")
        if elapsed > 0 and messages_received > 0:
            print(f"   Rate: {messages_received/elapsed:.1f} messages/second")
        print("="*60)

# Direct execution for Google Colab
print("🚀 Initializing WebSocket listener...")
print("This will run for 3 minutes and show live updates")
print("="*60)

# Run the async function directly
await listen_to_market_websocket(duration_minutes=3)
```

### Create Wallet and Setup Environment Variables

```python
def setup_wallet_and_env():
    """Create a new Ethereum wallet and setup environment variables"""

    # Create a new account
    account = Account.create()

    # Get wallet details
    private_key = account.key.hex()
    address = account.address

    # Create timestamp and nonce for authentication
    timestamp = str(int(time.time()))
    nonce = "0"

    # Create the message for EIP-712 signing
    message = f"Sign in to Polymarket\nAddress: {address.lower()}\nTimestamp: {timestamp}\nNonce: {nonce}"

    # Sign the message
    message_encoded = encode_defunct(text=message)
    signed_message = account.sign_message(message_encoded)
    signature = signed_message.signature.hex()

    # Create .env file content
    env_content = f"""# Polymarket Whale Tracker Configuration
WALLET_ADDRESS={address}
PRIVATE_KEY={private_key}
POLY_ADDRESS={address}
POLY_SIGNATURE={signature}
POLY_TIMESTAMP={timestamp}
POLY_NONCE={nonce}

# API Endpoints
GAMMA_API_URL=https://gamma-api.polymarket.com
DATA_API_URL=https://data-api.polymarket.com
CLOB_API_URL=https://clob.polymarket.com
WS_CLOB_URL=wss://ws-subscriptions-clob.polymarket.com/ws/market

# Database Configuration
DB_PATH=/content/polymarket_whales.db

# Tracking Configuration
MIN_TRADE_USD=1000
MAX_HOLDERS_PER_MARKET=100
"""

    # Write .env file
    env_path = Path("/content/.env")
    env_path.write_text(env_content)

    print(f"✅ Wallet created successfully!")
    print(f"📋 Address: {address}")
    print(f"🔑 Private key saved to .env file")
    print(f"📁 Environment file created at: {env_path}")

    # Load environment variables
    load_dotenv(env_path)

    return {
        "address": address,
        "private_key": private_key,
        "signature": signature,
        "timestamp": timestamp,
        "nonce": nonce
    }

# Execute wallet setup
wallet_info = setup_wallet_and_env()
```

#### Setup SQLite Database

```python
def setup_database():
    """Create SQLite database with all required tables"""

    db_path = os.getenv("DB_PATH", "/content/polymarket_whales.db")
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Create Events table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS events (
        id TEXT PRIMARY KEY,
        ticker TEXT,
        slug TEXT UNIQUE,
        title TEXT,
        description TEXT,
        start_date TEXT,
        end_date TEXT,
        liquidity REAL,
        volume REAL,
        volume_24hr REAL,
        active BOOLEAN,
        closed BOOLEAN,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        updated_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """)

    # Create Markets table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS markets (
        id TEXT PRIMARY KEY,
        event_id TEXT,
        question TEXT,
        condition_id TEXT,
        slug TEXT,
        outcomes TEXT,
        outcome_prices TEXT,
        clob_token_ids TEXT,
        liquidity REAL,
        volume REAL,
        volume_24hr REAL,
        active BOOLEAN,
        closed BOOLEAN,
        last_trade_price REAL,
        best_bid REAL,
        best_ask REAL,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        updated_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (event_id) REFERENCES events(id)
    )
    """)

    # Create Tags table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS tags (
        id INTEGER PRIMARY KEY,
        label TEXT,
        slug TEXT UNIQUE,
        force_show BOOLEAN,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        updated_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """)

    # Create Event_Tags junction table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS event_tags (
        event_id TEXT,
        tag_id INTEGER,
        PRIMARY KEY (event_id, tag_id),
        FOREIGN KEY (event_id) REFERENCES events(id),
        FOREIGN KEY (tag_id) REFERENCES tags(id)
    )
    """)

    # Create Teams table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS teams (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT,
        abbreviation TEXT,
        league TEXT,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """)

    # Create Market Holders table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS market_holders (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        market_id TEXT,
        token_id TEXT,
        proxy_wallet TEXT,
        username TEXT,
        pseudonym TEXT,
        amount REAL,
        outcome_index INTEGER,
        bio TEXT,
        profile_image TEXT,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (market_id) REFERENCES markets(id)
    )
    """)

    # Create Transactions table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS transactions (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        proxy_wallet TEXT,
        timestamp INTEGER,
        condition_id TEXT,
        transaction_hash TEXT UNIQUE,
        type TEXT,
        side TEXT,
        size REAL,
        usdc_size REAL,
        price REAL,
        asset TEXT,
        outcome_index INTEGER,
        title TEXT,
        slug TEXT,
        outcome TEXT,
        username TEXT,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """)

    # Create WebSocket Messages table for real-time data
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS ws_messages (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        market_id TEXT,
        asset_id TEXT,
        message_type TEXT,
        data TEXT,
        timestamp INTEGER,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """)

    conn.commit()
    conn.close()

    print(f"✅ Database setup complete at: {db_path}")
    print("📊 Tables created: events, markets, tags, event_tags, teams, market_holders, transactions, ws_messages")

# Execute database setup
setup_database()
```

```python
def fetch_all_events():
    """Fetch all active events from Gamma API and store in database"""

    base_url = os.getenv("GAMMA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    all_events = []
    offset = 0
    limit = 100

    print("🔍 Fetching all active events from Gamma API...")

    while True:
        try:
            url = f"{base_url}/events"
            params = {
                "closed": "false",
                "limit": limit,
                "offset": offset,
                "order": "volume",
                "ascending": "false"
            }

            response = requests.get(url, params=params)
            response.raise_for_status()

            events = response.json()

            if not events:
                break

            for event in events:
                # Store event in database
                cursor.execute("""
                INSERT OR REPLACE INTO events
                (id, ticker, slug, title, description, start_date, end_date,
                 liquidity, volume, volume_24hr, active, closed)
                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                """, (
                    event.get('id'),
                    event.get('ticker'),
                    event.get('slug'),
                    event.get('title'),
                    event.get('description'),
                    event.get('startDate'),
                    event.get('endDate'),
                    event.get('liquidity'),
                    event.get('volume'),
                    event.get('volume24hr'),
                    event.get('active'),
                    event.get('closed')
                ))

                # Store tags
                if 'tags' in event:
                    for tag in event['tags']:
                        cursor.execute("""
                        INSERT OR REPLACE INTO tags (id, label, slug, force_show)
                        VALUES (?, ?, ?, ?)
                        """, (
                            tag.get('id'),
                            tag.get('label'),
                            tag.get('slug'),
                            tag.get('forceShow', False)
                        ))

                        # Link event to tag
                        cursor.execute("""
                        INSERT OR IGNORE INTO event_tags (event_id, tag_id)
                        VALUES (?, ?)
                        """, (event.get('id'), tag.get('id')))

                all_events.append(event)

            print(f"  📦 Fetched {len(events)} events (offset: {offset})")
            offset += limit

            if len(events) < limit:
                break

        except Exception as e:
            print(f"❌ Error fetching events: {e}")
            break

    conn.commit()
    conn.close()

    print(f"✅ Total events fetched and stored: {len(all_events)}")
    return all_events

# Execute event fetching
events = fetch_all_events()
```

```python
# Cell 29: Fetch Market Tags by Market ID
def fetch_market_tags():
    """Fetch tags for each market using the market tags endpoint"""

    base_url = os.getenv("GAMMA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get all market IDs
    cursor.execute("SELECT id, question FROM markets ORDER BY id")
    markets = cursor.fetchall()

    print(f"🏷️ Fetching tags for {len(markets)} markets...")
    print("="*60)

    total_tags = 0
    markets_with_tags = 0
    tag_relationships_created = 0

    for idx, (market_id, question) in enumerate(markets, 1):
        try:
            url = f"{base_url}/markets/{market_id}/tags"

            response = requests.get(url)

            if response.status_code == 200:
                tags = response.json()

                if tags:
                    markets_with_tags += 1
                    market_tag_count = 0

                    for tag in tags:
                        # Update or insert tag with all available fields
                        cursor.execute("""
                        INSERT OR REPLACE INTO tags_enhanced
                        (id, label, slug, force_show, is_carousel, published_at,
                         created_at, updated_at)
                        VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                        """, (
                            tag.get('id'),
                            tag.get('label'),
                            tag.get('slug'),
                            tag.get('forceShow'),
                            tag.get('isCarousel'),
                            tag.get('publishedAt'),
                            tag.get('createdAt'),
                            tag.get('updatedAt')
                        ))

                        # Create or update market-tag relationship
                        cursor.execute("""
                        INSERT OR REPLACE INTO market_tags (market_id, tag_id)
                        VALUES (?, ?)
                        """, (market_id, tag.get('id')))

                        market_tag_count += 1
                        tag_relationships_created += 1
                        total_tags += 1

                    if markets_with_tags == 1:
                        print(f"  ✓ Example: {question[:50]}")
                        print(f"     Tags: {market_tag_count}")
                        for tag in tags[:3]:  # Show first 3 tags as example
                            print(f"       - {tag.get('label')} ({tag.get('slug')})")

        except Exception as e:
            # Silently skip markets without accessible tags
            pass

        # Progress update
        if idx % 250 == 0 or idx == len(markets):
            print(f"  📊 Progress: {idx}/{len(markets)} markets processed")
            print(f"     Markets with tags: {markets_with_tags}")
            print(f"     Total tag relationships: {tag_relationships_created}")
            conn.commit()

        # Check for milestones
        if tag_relationships_created % 1000 == 0 and tag_relationships_created > 0:
            print(f"  🎯 Milestone: {tag_relationships_created:,} tag relationships created!")

    # Get statistics
    cursor.execute("""
        SELECT COUNT(DISTINCT market_id) as markets_with_tags,
               COUNT(DISTINCT tag_id) as unique_tags,
               COUNT(*) as total_relationships
        FROM market_tags
    """)

    stats = cursor.fetchone()
    markets_with_tags_db, unique_tags_db, total_relationships_db = stats

    # Get most common tags for markets
    cursor.execute("""
        SELECT t.label, COUNT(mt.market_id) as market_count
        FROM market_tags mt
        JOIN tags_enhanced t ON mt.tag_id = t.id
        GROUP BY t.id, t.label
        ORDER BY market_count DESC
        LIMIT 10
    """)

    top_tags = cursor.fetchall()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Market tags fetch complete!")
    print(f"   Markets with tags: {markets_with_tags}/{len(markets)}")
    print(f"   Total tag relationships created: {tag_relationships_created}")
    print(f"\n📊 Database Statistics:")
    print(f"   Total markets with tags: {markets_with_tags_db}")
    print(f"   Unique tags used: {unique_tags_db}")
    print(f"   Total market-tag relationships: {total_relationships_db}")
    print(f"\n🏆 Top 10 Most Common Market Tags:")
    for tag_label, count in top_tags:
        print(f"   - {tag_label}: {count} markets")

# Execute market tags fetching
fetch_market_tags()
```

```python
# Cell 38: Fetch Current Positions for Users
def fetch_user_positions():
    """Fetch current positions for all users, especially whales"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get unique users from various sources
    cursor.execute("""
        SELECT DISTINCT proxy_wallet
        FROM (
            SELECT DISTINCT proxy_wallet FROM transactions WHERE usdc_size >= 100
            UNION
            SELECT DISTINCT proxy_wallet FROM market_holders WHERE amount >= 100
            UNION
            SELECT DISTINCT proxy_wallet FROM user_trades WHERE size >= 100
        )
        ORDER BY proxy_wallet
    """)

    users = cursor.fetchall()

    print(f"📊 Fetching current positions for {len(users)} users...")
    print("="*60)

    total_positions = 0
    users_with_positions = 0
    whale_positions = []

    for idx, (proxy_wallet,) in enumerate(users, 1):
        try:
            url = f"{data_api_url}/positions"
            params = {
                "user": proxy_wallet,
                "sizeThreshold": "1",
                "limit": "100",
                "sortBy": "TOKENS",
                "sortDirection": "DESC"
            }

            response = requests.get(url, params=params)

            if response.status_code == 200:
                positions = response.json()

                if positions:
                    users_with_positions += 1
                    user_total_value = 0

                    for position in positions:
                        # Store position
                        cursor.execute("""
                        INSERT OR REPLACE INTO user_positions_current
                        (proxy_wallet, asset, condition_id, size, avg_price,
                         initial_value, current_value, cash_pnl, percent_pnl,
                         total_bought, realized_pnl, percent_realized_pnl,
                         cur_price, redeemable, mergeable, negative_risk,
                         title, slug, icon, event_id, event_slug,
                         outcome, outcome_index, opposite_outcome, opposite_asset, end_date)
                        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                        """, (
                            position.get('proxyWallet'),
                            position.get('asset'),
                            position.get('conditionId'),
                            position.get('size'),
                            position.get('avgPrice'),
                            position.get('initialValue'),
                            position.get('currentValue'),
                            position.get('cashPnl'),
                            position.get('percentPnl'),
                            position.get('totalBought'),
                            position.get('realizedPnl'),
                            position.get('percentRealizedPnl'),
                            position.get('curPrice'),
                            position.get('redeemable'),
                            position.get('mergeable'),
                            position.get('negativeRisk'),
                            position.get('title'),
                            position.get('slug'),
                            position.get('icon'),
                            position.get('eventId'),
                            position.get('eventSlug'),
                            position.get('outcome'),
                            position.get('outcomeIndex'),
                            position.get('oppositeOutcome'),
                            position.get('oppositeAsset'),
                            position.get('endDate')
                        ))

                        user_total_value += position.get('currentValue', 0)
                        total_positions += 1

                        # Track whale positions (>$10k value)
                        if position.get('currentValue', 0) > 10000:
                            whale_positions.append({
                                'wallet': proxy_wallet,
                                'title': position.get('title'),
                                'value': position.get('currentValue'),
                                'pnl': position.get('percentPnl'),
                                'outcome': position.get('outcome')
                            })

                    # Update user total value
                    cursor.execute("""
                    INSERT OR REPLACE INTO users (proxy_wallet, total_value, last_updated)
                    VALUES (?, ?, CURRENT_TIMESTAMP)
                    ON CONFLICT(proxy_wallet) DO UPDATE SET
                        total_value = ?, last_updated = CURRENT_TIMESTAMP
                    """, (proxy_wallet, user_total_value, user_total_value))

                    if idx == 1:
                        print(f"  ✓ Example: {proxy_wallet[:10]}...")
                        print(f"     Positions: {len(positions)}")
                        print(f"     Total Value: ${user_total_value:,.2f}")
                        if positions:
                            print(f"     Largest: {positions[0].get('title', 'Unknown')[:40]}")
                            print(f"              ${positions[0].get('currentValue', 0):,.2f}")

        except Exception as e:
            # Silently skip users with errors
            pass

        # Progress update
        if idx % 100 == 0 or idx == len(users):
            print(f"  📊 Progress: {idx}/{len(users)} users processed")
            print(f"     Users with positions: {users_with_positions}")
            print(f"     Total positions: {total_positions}")
            conn.commit()

    # Sort whale positions by value
    whale_positions.sort(key=lambda x: x['value'], reverse=True)

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(DISTINCT proxy_wallet) as users_with_positions,
            COUNT(*) as total_positions,
            SUM(current_value) as total_value_locked,
            AVG(current_value) as avg_position_value,
            MAX(current_value) as largest_position,
            SUM(cash_pnl) as total_pnl,
            AVG(percent_pnl) as avg_percent_pnl
        FROM user_positions_current
    """)

    stats = cursor.fetchone()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ User positions fetch complete!")
    print(f"   Users processed: {len(users)}")
    print(f"   Users with positions: {users_with_positions}")
    print(f"   Total positions: {total_positions}")
    print(f"\n📊 Position Statistics:")
    if stats:
        users_count, pos_count, tvl, avg_val, largest, total_pnl, avg_pnl = stats
        print(f"   Total Value Locked: ${tvl or 0:,.2f}")
        print(f"   Average Position: ${avg_val or 0:,.2f}")
        print(f"   Largest Position: ${largest or 0:,.2f}")
        print(f"   Total P&L: ${total_pnl or 0:,.2f}")
        print(f"   Average P&L: {avg_pnl or 0:.2%}")

    print(f"\n🐋 Top 10 Whale Positions (>$10k):")
    for pos in whale_positions[:10]:
        print(f"   {pos['wallet'][:10]}... - {pos['title'][:40]}")
        print(f"     Value: ${pos['value']:,.2f} | P&L: {pos['pnl']:.2%} | {pos['outcome']}")

# Execute position fetching
fetch_user_positions()
```

```python
# Cell 39: Fetch Trades for Users and Markets
def fetch_user_trades():
    """Fetch trading activity for users and specific markets"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get users to fetch trades for (focusing on active traders)
    cursor.execute("""
        SELECT DISTINCT proxy_wallet
        FROM (
            SELECT proxy_wallet FROM user_positions_current
            UNION
            SELECT DISTINCT proxy_wallet FROM transactions WHERE usdc_size >= 100
        )
        LIMIT 500
    """)

    users = cursor.fetchall()

    print(f"📈 Fetching trades for {len(users)} active users...")
    print("="*60)

    total_trades = 0
    whale_trades = []

    for idx, (proxy_wallet,) in enumerate(users, 1):
        try:
            url = f"{data_api_url}/trades"
            params = {
                "user": proxy_wallet,
                "limit": "100",
                "takerOnly": "false"
            }

            response = requests.get(url, params=params)

            if response.status_code == 200:
                trades = response.json()

                for trade in trades:
                    # Calculate trade value
                    trade_value = trade.get('size', 0) * trade.get('price', 0)

                    # Store trade
                    cursor.execute("""
                    INSERT OR IGNORE INTO user_trades
                    (proxy_wallet, side, asset, condition_id, size, price,
                     timestamp, transaction_hash, title, slug, icon, event_slug,
                     outcome, outcome_index, username, pseudonym, bio,
                     profile_image, profile_image_optimized)
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                    """, (
                        trade.get('proxyWallet'),
                        trade.get('side'),
                        trade.get('asset'),
                        trade.get('conditionId'),
                        trade.get('size'),
                        trade.get('price'),
                        trade.get('timestamp'),
                        trade.get('transactionHash'),
                        trade.get('title'),
                        trade.get('slug'),
                        trade.get('icon'),
                        trade.get('eventSlug'),
                        trade.get('outcome'),
                        trade.get('outcomeIndex'),
                        trade.get('name'),
                        trade.get('pseudonym'),
                        trade.get('bio'),
                        trade.get('profileImage'),
                        trade.get('profileImageOptimized')
                    ))

                    # Update user info
                    if trade.get('name'):
                        cursor.execute("""
                        INSERT OR REPLACE INTO users
                        (proxy_wallet, username, pseudonym, bio, profile_image, profile_image_optimized)
                        VALUES (?, ?, ?, ?, ?, ?)
                        ON CONFLICT(proxy_wallet) DO UPDATE SET
                            username = COALESCE(?, username),
                            pseudonym = COALESCE(?, pseudonym),
                            bio = COALESCE(?, bio),
                            profile_image = COALESCE(?, profile_image),
                            profile_image_optimized = COALESCE(?, profile_image_optimized)
                        """, (
                            proxy_wallet,
                            trade.get('name'),
                            trade.get('pseudonym'),
                            trade.get('bio'),
                            trade.get('profileImage'),
                            trade.get('profileImageOptimized'),
                            trade.get('name'),
                            trade.get('pseudonym'),
                            trade.get('bio'),
                            trade.get('profileImage'),
                            trade.get('profileImageOptimized')
                        ))

                    total_trades += 1

                    # Track whale trades (>$10k value)
                    if trade_value > 10000:
                        whale_trades.append({
                            'wallet': proxy_wallet,
                            'title': trade.get('title'),
                            'value': trade_value,
                            'side': trade.get('side'),
                            'price': trade.get('price'),
                            'size': trade.get('size'),
                            'timestamp': trade.get('timestamp')
                        })

                if idx == 1 and trades:
                    print(f"  ✓ Example: {proxy_wallet[:10]}...")
                    print(f"     Trades: {len(trades)}")
                    print(f"     Latest: {trades[0].get('title', 'Unknown')[:40]}")
                    print(f"             {trades[0].get('side')} {trades[0].get('size', 0):.2f} @ {trades[0].get('price', 0):.4f}")

        except Exception as e:
            # Silently skip errors
            pass

        # Progress update
        if idx % 50 == 0 or idx == len(users):
            print(f"  📊 Progress: {idx}/{len(users)} users processed")
            print(f"     Total trades: {total_trades}")
            conn.commit()

    # Now fetch trades for top markets
    print(f"\n📈 Fetching trades for top markets...")

    cursor.execute("""
        SELECT DISTINCT condition_id, question
        FROM markets_enhanced
        WHERE active = 1 AND volume > 100000
        ORDER BY volume DESC
        LIMIT 50
    """)

    markets = cursor.fetchall()
    market_trades = 0

    for idx, (condition_id, question) in enumerate(markets, 1):
        try:
            url = f"{data_api_url}/trades"
            params = {
                "market": condition_id,
                "limit": "100"
            }

            response = requests.get(url, params=params)

            if response.status_code == 200:
                trades = response.json()

                for trade in trades:
                    # Store trade (using INSERT OR IGNORE to avoid duplicates)
                    cursor.execute("""
                    INSERT OR IGNORE INTO user_trades
                    (proxy_wallet, side, asset, condition_id, size, price,
                     timestamp, transaction_hash, title, slug, icon, event_slug,
                     outcome, outcome_index, username, pseudonym, bio,
                     profile_image, profile_image_optimized)
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                    """, (
                        trade.get('proxyWallet'),
                        trade.get('side'),
                        trade.get('asset'),
                        trade.get('conditionId'),
                        trade.get('size'),
                        trade.get('price'),
                        trade.get('timestamp'),
                        trade.get('transactionHash'),
                        trade.get('title'),
                        trade.get('slug'),
                        trade.get('icon'),
                        trade.get('eventSlug'),
                        trade.get('outcome'),
                        trade.get('outcomeIndex'),
                        trade.get('name'),
                        trade.get('pseudonym'),
                        trade.get('bio'),
                        trade.get('profileImage'),
                        trade.get('profileImageOptimized')
                    ))

                    market_trades += 1

        except Exception as e:
            # Silently skip errors
            pass

        # Progress update
        if idx % 10 == 0 or idx == len(markets):
            print(f"  📊 Markets processed: {idx}/{len(markets)} | Market trades: {market_trades}")

    # Sort whale trades
    whale_trades.sort(key=lambda x: x['value'], reverse=True)

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(*) as total_trades,
            COUNT(DISTINCT proxy_wallet) as unique_traders,
            COUNT(DISTINCT condition_id) as unique_markets,
            SUM(size * price) as total_volume,
            AVG(size * price) as avg_trade_value,
            MAX(size * price) as largest_trade
        FROM user_trades
    """)

    stats = cursor.fetchone()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Trades fetch complete!")
    print(f"   Total trades fetched: {total_trades + market_trades}")
    print(f"   User trades: {total_trades}")
    print(f"   Market trades: {market_trades}")

    if stats:
        total, traders, markets, volume, avg_value, largest = stats
        print(f"\n📊 Trade Statistics:")
        print(f"   Total trades in DB: {total}")
        print(f"   Unique traders: {traders}")
        print(f"   Unique markets: {markets}")
        print(f"   Total volume: ${volume or 0:,.2f}")
        print(f"   Average trade: ${avg_value or 0:,.2f}")
        print(f"   Largest trade: ${largest or 0:,.2f}")

    print(f"\n🐋 Top 10 Whale Trades (>$10k):")
    for trade in whale_trades[:10]:
        print(f"   {trade['wallet'][:10]}... - {trade['title'][:40]}")
        print(f"     {trade['side']}: ${trade['value']:,.2f} ({trade['size']:.2f} @ {trade['price']:.4f})")

# Execute trade fetching
fetch_user_trades()
```

#### Fetch Top Holders for First Market

```python
def fetch_top_holders(market_id=None):
    """Fetch top holders for a specific market or the first active market"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")
    min_balance = 1
    limit = 100

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # If no market_id provided, get the first active market
    if not market_id:
        cursor.execute("""
        SELECT id, condition_id, question
        FROM markets
        WHERE active = 1 AND condition_id IS NOT NULL
        LIMIT 1
        """)

        market_data = cursor.fetchone()

        if not market_data:
            print("❌ No active markets found")
            return []

        market_id, condition_id, question = market_data
        print(f"📊 Fetching top holders for: {question}")
    else:
        cursor.execute("SELECT condition_id FROM markets WHERE id = ?", (market_id,))
        result = cursor.fetchone()
        condition_id = result[0] if result else None

    if not condition_id:
        print(f"❌ No condition ID found for market {market_id}")
        return []

    # Convert condition_id to market hash format if needed
    if not condition_id.startswith('0x'):
        # You may need to adjust this conversion based on actual format
        market_hash = condition_id
    else:
        market_hash = condition_id

    print(f"   Market ID: {market_id}")
    print(f"   Condition ID: {market_hash}")

    all_holders = []

    try:
        url = f"{data_api_url}/holders"
        params = {
            "market": market_hash,
            "minBalance": min_balance,
            "limit": limit
        }

        response = requests.get(url, params=params)
        response.raise_for_status()

        holders_data = response.json()

        for token_group in holders_data:
            token_id = token_group.get('token')
            holders = token_group.get('holders', [])

            for holder in holders:
                # Store holder in database
                cursor.execute("""
                INSERT OR REPLACE INTO market_holders
                (market_id, token_id, proxy_wallet, username, pseudonym,
                 amount, outcome_index, bio, profile_image)
                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
                """, (
                    market_id,
                    token_id,
                    holder.get('proxyWallet'),
                    holder.get('name'),
                    holder.get('pseudonym'),
                    holder.get('amount'),
                    holder.get('outcomeIndex'),
                    holder.get('bio'),
                    holder.get('profileImage')
                ))

                all_holders.append(holder)

        conn.commit()
        print(f"✅ Fetched and stored {len(all_holders)} holders")

    except Exception as e:
        print(f"❌ Error fetching holders: {e}")

    conn.close()
    return all_holders

# Execute holder fetching
holders = fetch_top_holders()
```

#### Fetch Transactions for Top Holders

```python
def fetch_holder_transactions(holders, min_usdc_size=1000):
    """Fetch transaction activity for holders with trades above threshold"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    print(f"💰 Fetching transactions for {len(holders)} holders (min: ${min_usdc_size})...")

    whale_transactions = []
    processed_wallets = set()

    for holder in holders:
        wallet = holder.get('proxyWallet')

        if not wallet or wallet in processed_wallets:
            continue

        processed_wallets.add(wallet)

        try:
            url = f"{data_api_url}/activity"
            params = {
                "user": wallet,
                "limit": 100,
                "sortBy": "TIMESTAMP",
                "sortDirection": "DESC"
            }

            response = requests.get(url, params=params)
            response.raise_for_status()

            activities = response.json()

            for activity in activities:
                usdc_size = activity.get('usdcSize', 0)

                # Only store large transactions
                if usdc_size >= min_usdc_size:
                    cursor.execute("""
                    INSERT OR IGNORE INTO transactions
                    (proxy_wallet, timestamp, condition_id, transaction_hash,
                     type, side, size, usdc_size, price, asset, outcome_index,
                     title, slug, outcome, username)
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                    """, (
                        wallet,
                        activity.get('timestamp'),
                        activity.get('conditionId'),
                        activity.get('transactionHash'),
                        activity.get('type'),
                        activity.get('side'),
                        activity.get('size'),
                        usdc_size,
                        activity.get('price'),
                        activity.get('asset'),
                        activity.get('outcomeIndex'),
                        activity.get('title'),
                        activity.get('slug'),
                        activity.get('outcome'),
                        activity.get('name')
                    ))

                    whale_transactions.append(activity)

                    print(f"  🐋 WHALE ALERT: {activity.get('name', 'Unknown')} - "
                          f"${usdc_size:,.2f} {activity.get('side')} on {activity.get('title', 'Unknown')[:50]}")

        except Exception as e:
            print(f"  ⚠️ Error fetching transactions for {wallet}: {e}")

    conn.commit()
    conn.close()

    print(f"\n✅ Found {len(whale_transactions)} whale transactions (≥${min_usdc_size})")
    return whale_transactions

# Execute transaction fetching
whale_transactions = fetch_holder_transactions(holders)
```

#### Fetch User Portfolio Values

```python
# Cell 40: Fetch User Portfolio Values
def fetch_user_values():
    """Fetch total portfolio values for users and specific market positions"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get users with positions or significant trading activity
    cursor.execute("""
        SELECT DISTINCT proxy_wallet
        FROM (
            SELECT DISTINCT proxy_wallet FROM user_positions_current
            UNION
            SELECT DISTINCT proxy_wallet FROM user_trades
            GROUP BY proxy_wallet HAVING COUNT(*) > 5
        )
        ORDER BY proxy_wallet
    """)

    users = cursor.fetchall()

    print(f"💰 Fetching portfolio values for {len(users)} users...")
    print("="*60)

    total_value_fetched = 0
    whale_portfolios = []

    for idx, (proxy_wallet,) in enumerate(users, 1):
        try:
            # Fetch total portfolio value
            url = f"{data_api_url}/value"
            params = {"user": proxy_wallet}

            response = requests.get(url, params=params)

            if response.status_code == 200:
                value_data = response.json()

                if value_data and len(value_data) > 0:
                    total_value = value_data[0].get('value', 0)

                    # Store total value
                    cursor.execute("""
                    INSERT INTO user_values (proxy_wallet, market_condition_id, value)
                    VALUES (?, NULL, ?)
                    """, (proxy_wallet, total_value))

                    # Update user record
                    cursor.execute("""
                    UPDATE users SET total_value = ?, last_updated = CURRENT_TIMESTAMP
                    WHERE proxy_wallet = ?
                    """, (total_value, proxy_wallet))

                    total_value_fetched += 1

                    # Track whale portfolios (>$50k)
                    if total_value > 50000:
                        whale_portfolios.append({
                            'wallet': proxy_wallet,
                            'value': total_value
                        })

                    if idx == 1:
                        print(f"  ✓ Example: {proxy_wallet[:10]}...")
                        print(f"     Total Portfolio Value: ${total_value:,.2f}")

                    # Now fetch values for specific high-volume markets for this user
                    cursor.execute("""
                        SELECT DISTINCT condition_id
                        FROM user_positions_current
                        WHERE proxy_wallet = ? AND current_value > 1000
                        LIMIT 5
                    """, (proxy_wallet,))

                    user_markets = cursor.fetchall()

                    for market_id, in user_markets:
                        try:
                            params = {
                                "user": proxy_wallet,
                                "market": market_id
                            }

                            response = requests.get(url, params=params)

                            if response.status_code == 200:
                                market_value_data = response.json()

                                if market_value_data and len(market_value_data) > 0:
                                    market_value = market_value_data[0].get('value', 0)

                                    # Store market-specific value
                                    cursor.execute("""
                                    INSERT INTO user_values (proxy_wallet, market_condition_id, value)
                                    VALUES (?, ?, ?)
                                    """, (proxy_wallet, market_id, market_value))
                        except:
                            pass

        except Exception as e:
            # Silently skip errors
            pass

        # Progress update
        if idx % 100 == 0 or idx == len(users):
            print(f"  📊 Progress: {idx}/{len(users)} users processed")
            print(f"     Values fetched: {total_value_fetched}")
            print(f"     Whales found (>$50k): {len(whale_portfolios)}")
            conn.commit()

    # Sort whale portfolios
    whale_portfolios.sort(key=lambda x: x['value'], reverse=True)

    # Get portfolio statistics
    cursor.execute("""
        SELECT
            COUNT(DISTINCT proxy_wallet) as users_with_values,
            SUM(CASE WHEN market_condition_id IS NULL THEN value ELSE 0 END) as total_portfolio_value,
            AVG(CASE WHEN market_condition_id IS NULL THEN value ELSE 0 END) as avg_portfolio_value,
            MAX(CASE WHEN market_condition_id IS NULL THEN value ELSE 0 END) as largest_portfolio,
            COUNT(CASE WHEN market_condition_id IS NULL AND value > 50000 THEN 1 END) as whale_count,
            COUNT(CASE WHEN market_condition_id IS NULL AND value > 10000 THEN 1 END) as large_traders
        FROM user_values
    """)

    stats = cursor.fetchone()

    # Get distribution
    cursor.execute("""
        SELECT
            COUNT(CASE WHEN value < 100 THEN 1 END) as tiny,
            COUNT(CASE WHEN value >= 100 AND value < 1000 THEN 1 END) as small,
            COUNT(CASE WHEN value >= 1000 AND value < 10000 THEN 1 END) as medium,
            COUNT(CASE WHEN value >= 10000 AND value < 50000 THEN 1 END) as large,
            COUNT(CASE WHEN value >= 50000 THEN 1 END) as whale
        FROM user_values
        WHERE market_condition_id IS NULL
    """)

    distribution = cursor.fetchone()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Portfolio values fetch complete!")
    print(f"   Users processed: {len(users)}")
    print(f"   Values fetched: {total_value_fetched}")

    if stats:
        users_count, total_val, avg_val, largest, whales, large = stats
        print(f"\n📊 Portfolio Statistics:")
        print(f"   Users with values: {users_count}")
        print(f"   Total portfolio value: ${total_val or 0:,.2f}")
        print(f"   Average portfolio: ${avg_val or 0:,.2f}")
        print(f"   Largest portfolio: ${largest or 0:,.2f}")
        print(f"   Whales (>$50k): {whales}")
        print(f"   Large traders (>$10k): {large}")

    if distribution:
        tiny, small, medium, large, whale = distribution
        print(f"\n📈 Portfolio Distribution:")
        print(f"   <$100: {tiny} users")
        print(f"   $100-$1k: {small} users")
        print(f"   $1k-$10k: {medium} users")
        print(f"   $10k-$50k: {large} users")
        print(f"   >$50k: {whale} whales")

    print(f"\n🐋 Top 20 Whale Portfolios:")
    for i, portfolio in enumerate(whale_portfolios[:20], 1):
        print(f"   {i}. {portfolio['wallet'][:10]}... - ${portfolio['value']:,.2f}")

# Execute value fetching
fetch_user_values()
```

#### Fetch Closed Positions for Users

```python
# Cell 41: Fetch Closed Positions for Users
def fetch_closed_positions():
    """Fetch closed positions with realized P&L for users"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get users to check for closed positions
    cursor.execute("""
        SELECT DISTINCT proxy_wallet
        FROM (
            SELECT DISTINCT proxy_wallet FROM user_trades
            UNION
            SELECT DISTINCT proxy_wallet FROM user_positions_current
        )
        ORDER BY proxy_wallet
        LIMIT 1000
    """)

    users = cursor.fetchall()

    print(f"💸 Fetching closed positions for {len(users)} users...")
    print("="*60)

    total_closed_positions = 0
    users_with_closed = 0
    big_winners = []
    big_losers = []

    for idx, (proxy_wallet,) in enumerate(users, 1):
        try:
            url = f"{data_api_url}/closed-positions"
            params = {
                "user": proxy_wallet,
                "limit": "50",
                "sortBy": "REALIZEDPNL",
                "sortDirection": "DESC"
            }

            response = requests.get(url, params=params)

            if response.status_code == 200:
                closed_positions = response.json()

                if closed_positions:
                    users_with_closed += 1
                    user_total_pnl = 0

                    for position in closed_positions:
                        # Store closed position
                        cursor.execute("""
                        INSERT OR IGNORE INTO user_positions_closed
                        (proxy_wallet, asset, condition_id, avg_price,
                         total_bought, realized_pnl, cur_price,
                         title, slug, icon, event_slug,
                         outcome, outcome_index, opposite_outcome,
                         opposite_asset, end_date)
                        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                        """, (
                            position.get('proxyWallet'),
                            position.get('asset'),
                            position.get('conditionId'),
                            position.get('avgPrice'),
                            position.get('totalBought'),
                            position.get('realizedPnl'),
                            position.get('curPrice'),
                            position.get('title'),
                            position.get('slug'),
                            position.get('icon'),
                            position.get('eventSlug'),
                            position.get('outcome'),
                            position.get('outcomeIndex'),
                            position.get('oppositeOutcome'),
                            position.get('oppositeAsset'),
                            position.get('endDate')
                        ))

                        realized_pnl = position.get('realizedPnl', 0)
                        user_total_pnl += realized_pnl
                        total_closed_positions += 1

                        # Track big winners and losers
                        if realized_pnl > 5000:
                            big_winners.append({
                                'wallet': proxy_wallet,
                                'title': position.get('title'),
                                'pnl': realized_pnl,
                                'outcome': position.get('outcome'),
                                'avg_price': position.get('avgPrice'),
                                'cur_price': position.get('curPrice')
                            })
                        elif realized_pnl < -5000:
                            big_losers.append({
                                'wallet': proxy_wallet,
                                'title': position.get('title'),
                                'pnl': realized_pnl,
                                'outcome': position.get('outcome'),
                                'avg_price': position.get('avgPrice'),
                                'cur_price': position.get('curPrice')
                            })

                    if idx == 1:
                        print(f"  ✓ Example: {proxy_wallet[:10]}...")
                        print(f"     Closed positions: {len(closed_positions)}")
                        print(f"     Total realized P&L: ${user_total_pnl:,.2f}")
                        if closed_positions:
                            print(f"     Best trade: {closed_positions[0].get('title', 'Unknown')[:40]}")
                            print(f"                P&L: ${closed_positions[0].get('realizedPnl', 0):,.2f}")

            # Also fetch with specific market filters for high-volume markets
            if idx <= 10:  # Only for first few users to avoid excessive API calls
                cursor.execute("""
                    SELECT DISTINCT condition_id
                    FROM markets_enhanced
                    WHERE active = 0 AND volume > 500000
                    LIMIT 3
                """)

                high_volume_markets = cursor.fetchall()

                for market_id, in high_volume_markets:
                    try:
                        params = {
                            "user": proxy_wallet,
                            "market": market_id,
                            "limit": "10"
                        }

                        response = requests.get(url, params=params)

                        if response.status_code == 200:
                            market_closed = response.json()

                            for position in market_closed:
                                # Store if not already stored
                                cursor.execute("""
                                INSERT OR IGNORE INTO user_positions_closed
                                (proxy_wallet, asset, condition_id, avg_price,
                                 total_bought, realized_pnl, cur_price,
                                 title, slug, icon, event_slug,
                                 outcome, outcome_index, opposite_outcome,
                                 opposite_asset, end_date)
                                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                                """, (
                                    position.get('proxyWallet'),
                                    position.get('asset'),
                                    position.get('conditionId'),
                                    position.get('avgPrice'),
                                    position.get('totalBought'),
                                    position.get('realizedPnl'),
                                    position.get('curPrice'),
                                    position.get('title'),
                                    position.get('slug'),
                                    position.get('icon'),
                                    position.get('eventSlug'),
                                    position.get('outcome'),
                                    position.get('outcomeIndex'),
                                    position.get('oppositeOutcome'),
                                    position.get('oppositeAsset'),
                                    position.get('endDate')
                                ))
                    except:
                        pass

        except Exception as e:
            # Silently skip errors
            pass

        # Progress update
        if idx % 100 == 0 or idx == len(users):
            print(f"  📊 Progress: {idx}/{len(users)} users processed")
            print(f"     Users with closed positions: {users_with_closed}")
            print(f"     Total closed positions: {total_closed_positions}")
            conn.commit()

    # Sort winners and losers
    big_winners.sort(key=lambda x: x['pnl'], reverse=True)
    big_losers.sort(key=lambda x: x['pnl'])

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(DISTINCT proxy_wallet) as users_with_closed,
            COUNT(*) as total_positions,
            SUM(realized_pnl) as total_realized_pnl,
            AVG(realized_pnl) as avg_realized_pnl,
            MAX(realized_pnl) as best_trade,
            MIN(realized_pnl) as worst_trade,
            SUM(CASE WHEN realized_pnl > 0 THEN 1 ELSE 0 END) as winning_trades,
            SUM(CASE WHEN realized_pnl < 0 THEN 1 ELSE 0 END) as losing_trades,
            SUM(CASE WHEN realized_pnl > 0 THEN realized_pnl ELSE 0 END) as total_wins,
            SUM(CASE WHEN realized_pnl < 0 THEN realized_pnl ELSE 0 END) as total_losses
        FROM user_positions_closed
    """)

    stats = cursor.fetchone()

    # Get top performers
    cursor.execute("""
        SELECT proxy_wallet, COUNT(*) as trades, SUM(realized_pnl) as total_pnl
        FROM user_positions_closed
        GROUP BY proxy_wallet
        HAVING trades > 5
        ORDER BY total_pnl DESC
        LIMIT 10
    """)

    top_performers = cursor.fetchall()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Closed positions fetch complete!")
    print(f"   Users processed: {len(users)}")
    print(f"   Users with closed positions: {users_with_closed}")
    print(f"   Total closed positions: {total_closed_positions}")

    if stats:
        (users_count, total_pos, total_pnl, avg_pnl, best, worst,
         wins, losses, total_win_amt, total_loss_amt) = stats

        print(f"\n📊 Closed Position Statistics:")
        print(f"   Total positions: {total_pos}")
        print(f"   Total realized P&L: ${total_pnl or 0:,.2f}")
        print(f"   Average P&L per trade: ${avg_pnl or 0:,.2f}")
        print(f"   Best trade: ${best or 0:,.2f}")
        print(f"   Worst trade: ${worst or 0:,.2f}")
        print(f"   Win rate: {wins/(wins+losses)*100 if (wins+losses) > 0 else 0:.1f}% ({wins} wins / {losses} losses)")
        print(f"   Total wins: ${total_win_amt or 0:,.2f}")
        print(f"   Total losses: ${total_loss_amt or 0:,.2f}")

    print(f"\n🏆 Top 10 Profitable Traders (>5 trades):")
    for wallet, trades, pnl in top_performers:
        print(f"   {wallet[:10]}... - {trades} trades | P&L: ${pnl:,.2f}")

    print(f"\n💰 Top 10 Winning Trades (>$5k profit):")
    for win in big_winners[:10]:
        print(f"   {win['wallet'][:10]}... - {win['title'][:40]}")
        print(f"     P&L: ${win['pnl']:,.2f} | {win['outcome']} | Buy: {win['avg_price']:.4f} → Sell: {win['cur_price']:.4f}")

    print(f"\n💸 Top 10 Losing Trades (>$5k loss):")
    for loss in big_losers[:10]:
        print(f"   {loss['wallet'][:10]}... - {loss['title'][:40]}")
        print(f"     P&L: ${loss['pnl']:,.2f} | {loss['outcome']} | Buy: {loss['avg_price']:.4f} → End: {loss['cur_price']:.4f}")

# Execute closed positions fetching
fetch_closed_positions()
```

```python
# Cell 42: Fetch User Trading Statistics
def fetch_user_trading_stats():
    """Fetch total markets traded for all users"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Add trading stats columns to users table if not exists
    cursor.execute("""
    ALTER TABLE users ADD COLUMN markets_traded INTEGER DEFAULT 0
    """)
    conn.commit()

    # Get all users with trading activity
    cursor.execute("""
        SELECT DISTINCT proxy_wallet, username
        FROM users
        WHERE proxy_wallet IN (
            SELECT DISTINCT proxy_wallet FROM user_trades
            UNION
            SELECT DISTINCT proxy_wallet FROM user_positions_current
        )
        ORDER BY proxy_wallet
    """)

    users = cursor.fetchall()

    print(f"📊 Fetching trading statistics for {len(users)} users...")
    print("="*60)

    total_fetched = 0
    active_traders = []

    for idx, (proxy_wallet, username) in enumerate(users, 1):
        try:
            url = f"{data_api_url}/traded"
            params = {"user": proxy_wallet}

            response = requests.get(url, params=params)

            if response.status_code == 200:
                data = response.json()

                if data:
                    markets_traded = data.get('traded', 0)

                    # Update user record
                    cursor.execute("""
                    UPDATE users
                    SET markets_traded = ?, last_updated = CURRENT_TIMESTAMP
                    WHERE proxy_wallet = ?
                    """, (markets_traded, proxy_wallet))

                    total_fetched += 1

                    # Track very active traders (>100 markets)
                    if markets_traded > 100:
                        active_traders.append({
                            'wallet': proxy_wallet,
                            'username': username,
                            'markets': markets_traded
                        })

                    if idx == 1:
                        print(f"  ✓ Example: {username or proxy_wallet[:10]}...")
                        print(f"     Markets traded: {markets_traded}")

        except Exception as e:
            # Silently skip errors
            pass

        # Progress update
        if idx % 100 == 0 or idx == len(users):
            print(f"  📊 Progress: {idx}/{len(users)} users processed")
            print(f"     Stats fetched: {total_fetched}")
            conn.commit()

    # Sort active traders
    active_traders.sort(key=lambda x: x['markets'], reverse=True)

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(*) as total_users,
            SUM(markets_traded) as total_market_interactions,
            AVG(markets_traded) as avg_markets_per_user,
            MAX(markets_traded) as most_active_user_markets,
            COUNT(CASE WHEN markets_traded > 10 THEN 1 END) as active_traders,
            COUNT(CASE WHEN markets_traded > 100 THEN 1 END) as very_active_traders,
            COUNT(CASE WHEN markets_traded > 1000 THEN 1 END) as super_active_traders
        FROM users
        WHERE markets_traded > 0
    """)

    stats = cursor.fetchone()

    # Get distribution
    cursor.execute("""
        SELECT
            COUNT(CASE WHEN markets_traded = 1 THEN 1 END) as single_market,
            COUNT(CASE WHEN markets_traded BETWEEN 2 AND 5 THEN 1 END) as few_markets,
            COUNT(CASE WHEN markets_traded BETWEEN 6 AND 20 THEN 1 END) as moderate,
            COUNT(CASE WHEN markets_traded BETWEEN 21 AND 100 THEN 1 END) as active,
            COUNT(CASE WHEN markets_traded > 100 THEN 1 END) as very_active
        FROM users
        WHERE markets_traded > 0
    """)

    distribution = cursor.fetchone()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Trading statistics fetch complete!")
    print(f"   Users processed: {len(users)}")
    print(f"   Stats fetched: {total_fetched}")

    if stats:
        (total, interactions, avg, most, active, very_active, super_active) = stats
        print(f"\n📊 User Trading Statistics:")
        print(f"   Total traders: {total}")
        print(f"   Total market interactions: {interactions}")
        print(f"   Average markets per user: {avg:.1f}")
        print(f"   Most active user: {most} markets")
        print(f"   Active traders (>10 markets): {active}")
        print(f"   Very active (>100 markets): {very_active}")
        print(f"   Super active (>1000 markets): {super_active}")

    if distribution:
        single, few, moderate, active, very = distribution
        print(f"\n📈 Trading Activity Distribution:")
        print(f"   1 market only: {single} users")
        print(f"   2-5 markets: {few} users")
        print(f"   6-20 markets: {moderate} users")
        print(f"   21-100 markets: {active} users")
        print(f"   >100 markets: {very} users")

    print(f"\n🏆 Top 20 Most Active Traders:")
    for i, trader in enumerate(active_traders[:20], 1):
        display_name = trader['username'] or trader['wallet'][:10] + "..."
        print(f"   {i}. {display_name}: {trader['markets']} markets")

# Execute trading stats fetching
fetch_user_trading_stats()
```

#### Fetch Open Interest for Markets

```python
# Cell 43: Fetch Open Interest for Markets
def fetch_open_interest():
    """Fetch open interest data for markets"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Add open interest table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS market_open_interest (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        market_id TEXT,
        condition_id TEXT,
        open_interest REAL,
        timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (market_id) REFERENCES markets(id)
    )
    """)

    # Get all active markets
    cursor.execute("""
        SELECT id, condition_id, question, volume
        FROM markets_enhanced
        WHERE active = 1
        ORDER BY volume DESC
        LIMIT 500
    """)

    markets = cursor.fetchall()

    print(f"💰 Fetching open interest for {len(markets)} markets...")
    print("="*60)

    total_fetched = 0
    high_oi_markets = []
    global_oi = 0

    # First, get global open interest
    try:
        url = f"{data_api_url}/oi"
        params = {"market": "GLOBAL"}

        response = requests.get(url, params=params)

        if response.status_code == 200:
            data = response.json()

            if data and len(data) > 0:
                global_oi = data[0].get('value', 0)

                # Store global OI
                cursor.execute("""
                INSERT INTO market_open_interest (market_id, condition_id, open_interest)
                VALUES ('GLOBAL', 'GLOBAL', ?)
                """, (global_oi,))

                print(f"  ✓ Global Open Interest: ${global_oi:,.2f}")
    except:
        pass

    print(f"\n📊 Fetching individual market open interest...")

    for idx, (market_id, condition_id, question, volume) in enumerate(markets, 1):
        try:
            # Try with market ID first
            url = f"{data_api_url}/oi"
            params = {"market": market_id}

            response = requests.get(url, params=params)

            if response.status_code == 200:
                data = response.json()

                if data and len(data) > 0:
                    oi_value = data[0].get('value', 0)

                    # Store open interest
                    cursor.execute("""
                    INSERT INTO market_open_interest (market_id, condition_id, open_interest)
                    VALUES (?, ?, ?)
                    """, (market_id, condition_id, oi_value))

                    # Update markets table
                    cursor.execute("""
                    UPDATE markets_enhanced
                    SET open_interest = ?
                    WHERE id = ?
                    """, (oi_value, market_id))

                    total_fetched += 1

                    # Track high open interest markets (>$100k)
                    if oi_value > 100000:
                        high_oi_markets.append({
                            'id': market_id,
                            'question': question,
                            'oi': oi_value,
                            'volume': volume
                        })

                    if idx == 1:
                        print(f"  ✓ Example: {question[:50]}...")
                        print(f"     Open Interest: ${oi_value:,.2f}")
                        print(f"     Volume: ${volume:,.2f}")
                        if volume > 0:
                            print(f"     OI/Volume Ratio: {oi_value/volume:.2f}")

        except Exception as e:
            # Try with condition_id if market_id fails
            try:
                params = {"market": condition_id}
                response = requests.get(url, params=params)

                if response.status_code == 200:
                    data = response.json()

                    if data and len(data) > 0:
                        oi_value = data[0].get('value', 0)

                        cursor.execute("""
                        INSERT INTO market_open_interest (market_id, condition_id, open_interest)
                        VALUES (?, ?, ?)
                        """, (market_id, condition_id, oi_value))

                        total_fetched += 1
            except:
                pass

        # Progress update
        if idx % 50 == 0 or idx == len(markets):
            print(f"  📊 Progress: {idx}/{len(markets)} markets processed")
            print(f"     Open interest data fetched: {total_fetched}")
            conn.commit()

    # Sort high OI markets
    high_oi_markets.sort(key=lambda x: x['oi'], reverse=True)

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(*) as markets_with_oi,
            SUM(open_interest) as total_oi,
            AVG(open_interest) as avg_oi,
            MAX(open_interest) as max_oi,
            MIN(open_interest) as min_oi,
            COUNT(CASE WHEN open_interest > 100000 THEN 1 END) as high_oi_markets,
            COUNT(CASE WHEN open_interest > 1000000 THEN 1 END) as very_high_oi_markets
        FROM market_open_interest
        WHERE market_id != 'GLOBAL'
        AND timestamp = (SELECT MAX(timestamp) FROM market_open_interest WHERE market_id != 'GLOBAL')
    """)

    stats = cursor.fetchone()

    # Get OI/Volume ratios
    cursor.execute("""
        SELECT
            m.question,
            moi.open_interest,
            m.volume,
            moi.open_interest / NULLIF(m.volume, 0) as oi_volume_ratio
        FROM market_open_interest moi
        JOIN markets_enhanced m ON moi.market_id = m.id
        WHERE m.volume > 10000
        AND moi.timestamp = (SELECT MAX(timestamp) FROM market_open_interest WHERE market_id = moi.market_id)
        ORDER BY oi_volume_ratio DESC
        LIMIT 10
    """)

    high_ratio_markets = cursor.fetchall()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Open interest fetch complete!")
    print(f"   Markets processed: {len(markets)}")
    print(f"   Open interest fetched: {total_fetched}")
    print(f"   Global Open Interest: ${global_oi:,.2f}")

    if stats:
        markets_count, total, avg, max_oi, min_oi, high, very_high = stats
        print(f"\n📊 Open Interest Statistics:")
        print(f"   Markets with OI data: {markets_count}")
        print(f"   Total Open Interest: ${total or 0:,.2f}")
        print(f"   Average OI per market: ${avg or 0:,.2f}")
        print(f"   Maximum OI: ${max_oi or 0:,.2f}")
        print(f"   High OI markets (>$100k): {high}")
        print(f"   Very high OI (>$1M): {very_high}")

    print(f"\n💰 Top 10 Markets by Open Interest:")
    for i, market in enumerate(high_oi_markets[:10], 1):
        print(f"   {i}. {market['question'][:50]}...")
        print(f"      Open Interest: ${market['oi']:,.2f}")
        print(f"      Volume: ${market['volume']:,.2f}")
        if market['volume'] > 0:
            print(f"      OI/Volume Ratio: {market['oi']/market['volume']:.2f}")

    print(f"\n📈 Top 10 Markets by OI/Volume Ratio (high interest retention):")
    for i, (question, oi, volume, ratio) in enumerate(high_ratio_markets, 1):
        print(f"   {i}. {question[:50]}...")
        print(f"      Ratio: {ratio:.2f} (OI: ${oi:,.2f} / Vol: ${volume:,.2f})")

# Execute open interest fetching
fetch_open_interest()
```

#### Fetch Live Volume for Events

```python
# Cell 44: Fetch Live Volume for Events
def fetch_live_volume():
    """Fetch live volume data for events"""

    data_api_url = os.getenv("DATA_API_URL")
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Create live volume table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS event_live_volume (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        event_id TEXT,
        total_volume REAL,
        market_volumes TEXT,  -- JSON string of market volumes
        timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (event_id) REFERENCES events(id)
    )
    """)

    # Get all active events
    cursor.execute("""
        SELECT id, title, volume
        FROM events_enhanced
        WHERE active = 1
        ORDER BY volume DESC
        LIMIT 200
    """)

    events = cursor.fetchall()

    print(f"📊 Fetching live volume for {len(events)} events...")
    print("="*60)

    total_fetched = 0
    high_volume_events = []
    total_live_volume = 0

    for idx, (event_id, title, stored_volume) in enumerate(events, 1):
        try:
            url = f"{data_api_url}/live-volume"
            params = {"id": event_id}

            response = requests.get(url, params=params)

            if response.status_code == 200:
                data = response.json()

                if data and len(data) > 0:
                    volume_data = data[0]
                    live_total = volume_data.get('total', 0)
                    markets_data = volume_data.get('markets', [])

                    # Convert markets data to JSON string for storage
                    markets_json = json.dumps(markets_data)

                    # Store live volume
                    cursor.execute("""
                    INSERT INTO event_live_volume (event_id, total_volume, market_volumes)
                    VALUES (?, ?, ?)
                    """, (event_id, live_total, markets_json))

                    # Update events table with latest volume
                    cursor.execute("""
                    UPDATE events_enhanced
                    SET volume = ?, updated_at = CURRENT_TIMESTAMP
                    WHERE id = ?
                    """, (live_total, event_id))

                    total_fetched += 1
                    total_live_volume += live_total

                    # Calculate volume change
                    volume_change = live_total - (stored_volume or 0) if stored_volume else 0

                    # Track high volume events (>$100k)
                    if live_total > 100000:
                        high_volume_events.append({
                            'id': event_id,
                            'title': title,
                            'live_volume': live_total,
                            'stored_volume': stored_volume,
                            'change': volume_change,
                            'markets_count': len(markets_data)
                        })

                    # Update individual market volumes
                    for market in markets_data:
                        market_id = market.get('market')
                        market_volume = market.get('value', 0)

                        cursor.execute("""
                        UPDATE markets_enhanced
                        SET volume = ?, updated_at = CURRENT_TIMESTAMP
                        WHERE condition_id = ?
                        """, (market_volume, market_id))

                    if idx == 1:
                        print(f"  ✓ Example: {title[:50]}...")
                        print(f"     Live Volume: ${live_total:,.2f}")
                        print(f"     Stored Volume: ${stored_volume or 0:,.2f}")
                        print(f"     Change: ${volume_change:,.2f} ({volume_change/(stored_volume or 1)*100:.1f}%)")
                        print(f"     Markets: {len(markets_data)}")
                        if markets_data:
                            print(f"     Largest market: ${max(m['value'] for m in markets_data):,.2f}")

        except Exception as e:
            # Silently skip errors
            pass

        # Progress update
        if idx % 20 == 0 or idx == len(events):
            print(f"  📊 Progress: {idx}/{len(events)} events processed")
            print(f"     Live volumes fetched: {total_fetched}")
            print(f"     Total live volume: ${total_live_volume:,.2f}")
            conn.commit()

    # Sort high volume events
    high_volume_events.sort(key=lambda x: x['live_volume'], reverse=True)

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(DISTINCT event_id) as events_with_volume,
            SUM(total_volume) as total_volume,
            AVG(total_volume) as avg_volume,
            MAX(total_volume) as max_volume,
            COUNT(CASE WHEN total_volume > 100000 THEN 1 END) as high_volume_events,
            COUNT(CASE WHEN total_volume > 1000000 THEN 1 END) as very_high_volume_events
        FROM event_live_volume
        WHERE timestamp = (SELECT MAX(timestamp) FROM event_live_volume)
    """)

    stats = cursor.fetchone()

    # Get events with most markets
    cursor.execute("""
        SELECT
            e.title,
            elv.event_id,
            elv.total_volume,
            LENGTH(elv.market_volumes) - LENGTH(REPLACE(elv.market_volumes, 'market', '')) as market_count
        FROM event_live_volume elv
        JOIN events_enhanced e ON elv.event_id = e.id
        WHERE elv.timestamp = (SELECT MAX(timestamp) FROM event_live_volume)
        ORDER BY market_count DESC
        LIMIT 10
    """)

    multi_market_events = cursor.fetchall()

    # Get events with biggest volume changes
    cursor.execute("""
        SELECT
            e.title,
            elv.total_volume as current_volume,
            e.volume as stored_volume,
            elv.total_volume - e.volume as volume_change,
            (elv.total_volume - e.volume) / NULLIF(e.volume, 0) * 100 as percent_change
        FROM event_live_volume elv
        JOIN events_enhanced e ON elv.event_id = e.id
        WHERE elv.timestamp = (SELECT MAX(timestamp) FROM event_live_volume)
        AND e.volume > 10000
        ORDER BY ABS(volume_change) DESC
        LIMIT 10
    """)

    volume_changes = cursor.fetchall()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Live volume fetch complete!")
    print(f"   Events processed: {len(events)}")
    print(f"   Live volumes fetched: {total_fetched}")
    print(f"   Total live volume: ${total_live_volume:,.2f}")

    if stats:
        events_count, total, avg, max_vol, high, very_high = stats
        print(f"\n📊 Live Volume Statistics:")
        print(f"   Events with volume: {events_count}")
        print(f"   Total volume: ${total or 0:,.2f}")
        print(f"   Average volume: ${avg or 0:,.2f}")
        print(f"   Maximum volume: ${max_vol or 0:,.2f}")
        print(f"   High volume events (>$100k): {high}")
        print(f"   Very high volume (>$1M): {very_high}")

    print(f"\n💰 Top 10 Events by Live Volume:")
    for i, event in enumerate(high_volume_events[:10], 1):
        print(f"   {i}. {event['title'][:50]}...")
        print(f"      Live: ${event['live_volume']:,.2f} | Stored: ${event['stored_volume'] or 0:,.2f}")
        print(f"      Change: ${event['change']:,.2f} | Markets: {event['markets_count']}")

    print(f"\n📈 Events with Most Markets:")
    for i, (title, event_id, volume, market_count) in enumerate(multi_market_events, 1):
        print(f"   {i}. {title[:50]}...")
        print(f"      Markets: ~{market_count//6} | Volume: ${volume:,.2f}")

    print(f"\n🔄 Biggest Volume Changes:")
    for i, (title, current, stored, change, percent) in enumerate(volume_changes, 1):
        symbol = "📈" if change > 0 else "📉"
        print(f"   {i}. {symbol} {title[:50]}...")
        print(f"      Change: ${change:,.2f} ({percent or 0:.1f}%)")
        print(f"      Current: ${current:,.2f} | Previous: ${stored or 0:,.2f}")

# Execute live volume fetching
fetch_live_volume()
```

```python
# Cell 45: Update Database Schema for CLOB Data
def update_clob_schema():
    """Create tables for CLOB order book and pricing data"""

    db_path = os.getenv("DB_PATH", "/content/polymarket_whales.db")
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Create order book table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS order_books (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        token_id TEXT,
        market TEXT,
        asset_id TEXT,
        timestamp TEXT,
        hash TEXT,
        bids TEXT,  -- JSON array of bid orders
        asks TEXT,  -- JSON array of ask orders
        min_order_size TEXT,
        tick_size TEXT,
        neg_risk BOOLEAN,
        best_bid REAL,
        best_ask REAL,
        spread REAL,
        total_bid_volume REAL,
        total_ask_volume REAL,
        fetched_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        UNIQUE(token_id, timestamp)
    )
    """)

    # Create price tracking table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS clob_prices (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        token_id TEXT,
        buy_price REAL,
        sell_price REAL,
        mid_price REAL,
        spread REAL,
        timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        UNIQUE(token_id, timestamp)
    )
    """)

    # Create price history table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS price_history (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        token_id TEXT,
        timestamp INTEGER,
        price REAL,
        fetched_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        UNIQUE(token_id, timestamp)
    )
    """)

    # Create spreads table
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS market_spreads (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        token_id TEXT,
        spread REAL,
        bid_price REAL,
        ask_price REAL,
        timestamp TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """)

    # Create indexes
    cursor.execute("CREATE INDEX IF NOT EXISTS idx_orderbooks_token ON order_books(token_id)")
    cursor.execute("CREATE INDEX IF NOT EXISTS idx_orderbooks_timestamp ON order_books(timestamp)")
    cursor.execute("CREATE INDEX IF NOT EXISTS idx_prices_token ON clob_prices(token_id)")
    cursor.execute("CREATE INDEX IF NOT EXISTS idx_history_token ON price_history(token_id)")
    cursor.execute("CREATE INDEX IF NOT EXISTS idx_history_timestamp ON price_history(timestamp)")
    cursor.execute("CREATE INDEX IF NOT EXISTS idx_spreads_token ON market_spreads(token_id)")

    conn.commit()
    conn.close()

    print("✅ CLOB database schema created successfully!")
    print("📊 Tables created:")
    print("   - order_books: Full order book snapshots")
    print("   - clob_prices: Current buy/sell prices")
    print("   - price_history: Historical price data")
    print("   - market_spreads: Bid-ask spread tracking")

# Execute schema creation
update_clob_schema()
```

```python
# Cell 46: Fetch Order Book Data from CLOB
def fetch_order_books():
    """Fetch order book data for all active market tokens"""

    clob_url = "https://clob.polymarket.com"
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get all token IDs from active markets
    cursor.execute("""
        SELECT DISTINCT
            m.clob_token_ids,
            m.condition_id,
            m.question
        FROM markets_enhanced m
        WHERE m.active = 1
        AND m.clob_token_ids IS NOT NULL
        AND m.clob_token_ids != '[]'
        ORDER BY m.volume DESC
        LIMIT 100
    """)

    markets = cursor.fetchall()

    print(f"📚 Fetching order books for {len(markets)} markets...")
    print("="*60)

    total_order_books = 0
    markets_with_liquidity = 0
    high_liquidity_markets = []

    for idx, (token_ids_str, condition_id, question) in enumerate(markets, 1):
        try:
            # Parse token IDs
            token_ids = json.loads(token_ids_str) if token_ids_str else []

            for token_id in token_ids[:2]:  # Get first 2 tokens (Yes/No typically)
                try:
                    url = f"{clob_url}/book"
                    params = {"token_id": token_id}

                    response = requests.get(url, params=params)

                    if response.status_code == 200:
                        book = response.json()

                        # Calculate order book metrics
                        bids = book.get('bids', [])
                        asks = book.get('asks', [])

                        best_bid = float(bids[0]['price']) if bids else 0
                        best_ask = float(asks[0]['price']) if asks else 1
                        spread = best_ask - best_bid if bids and asks else 1

                        # Calculate volumes
                        bid_volume = sum(float(b['size']) for b in bids)
                        ask_volume = sum(float(a['size']) for a in asks)
                        total_volume = bid_volume + ask_volume

                        # Store order book
                        cursor.execute("""
                        INSERT OR REPLACE INTO order_books
                        (token_id, market, asset_id, timestamp, hash,
                         bids, asks, min_order_size, tick_size, neg_risk,
                         best_bid, best_ask, spread, total_bid_volume, total_ask_volume)
                        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                        """, (
                            token_id,
                            book.get('market'),
                            book.get('asset_id'),
                            book.get('timestamp'),
                            book.get('hash'),
                            json.dumps(bids),
                            json.dumps(asks),
                            book.get('min_order_size'),
                            book.get('tick_size'),
                            book.get('neg_risk'),
                            best_bid,
                            best_ask,
                            spread,
                            bid_volume,
                            ask_volume
                        ))

                        total_order_books += 1

                        if total_volume > 1000:
                            markets_with_liquidity += 1

                            if total_volume > 10000:
                                high_liquidity_markets.append({
                                    'question': question,
                                    'token_id': token_id,
                                    'volume': total_volume,
                                    'spread': spread,
                                    'best_bid': best_bid,
                                    'best_ask': best_ask
                                })

                        if idx == 1 and token_id == token_ids[0]:
                            print(f"  ✓ Example: {question[:50]}...")
                            print(f"     Token: {token_id[:20]}...")
                            print(f"     Best Bid: {best_bid:.4f} | Best Ask: {best_ask:.4f}")
                            print(f"     Spread: {spread:.4f} ({spread*100:.2f}%)")
                            print(f"     Liquidity: {total_volume:.2f} shares")

                except Exception as e:
                    # Silently skip token errors
                    pass

        except Exception as e:
            # Silently skip market errors
            pass

        # Progress update
        if idx % 20 == 0 or idx == len(markets):
            print(f"  📊 Progress: {idx}/{len(markets)} markets processed")
            print(f"     Order books fetched: {total_order_books}")
            print(f"     Markets with liquidity: {markets_with_liquidity}")
            conn.commit()

    # Sort high liquidity markets
    high_liquidity_markets.sort(key=lambda x: x['volume'], reverse=True)

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(DISTINCT token_id) as unique_tokens,
            AVG(spread) as avg_spread,
            MIN(spread) as tightest_spread,
            MAX(spread) as widest_spread,
            AVG(total_bid_volume + total_ask_volume) as avg_liquidity,
            MAX(total_bid_volume + total_ask_volume) as max_liquidity
        FROM order_books
        WHERE fetched_at = (SELECT MAX(fetched_at) FROM order_books)
    """)

    stats = cursor.fetchone()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Order book fetch complete!")
    print(f"   Markets processed: {len(markets)}")
    print(f"   Order books fetched: {total_order_books}")
    print(f"   Markets with liquidity: {markets_with_liquidity}")

    if stats:
        tokens, avg_spread, min_spread, max_spread, avg_liq, max_liq = stats
        print(f"\n📊 Order Book Statistics:")
        print(f"   Unique tokens: {tokens}")
        print(f"   Average spread: {avg_spread or 0:.4f}")
        print(f"   Tightest spread: {min_spread or 0:.4f}")
        print(f"   Widest spread: {max_spread or 0:.4f}")
        print(f"   Average liquidity: {avg_liq or 0:.2f} shares")
        print(f"   Maximum liquidity: {max_liq or 0:.2f} shares")

    print(f"\n💰 Top 10 Markets by Liquidity:")
    for i, market in enumerate(high_liquidity_markets[:10], 1):
        print(f"   {i}. {market['question'][:50]}...")
        print(f"      Liquidity: {market['volume']:,.2f} shares")
        print(f"      Bid: {market['best_bid']:.4f} | Ask: {market['best_ask']:.4f} | Spread: {market['spread']:.4f}")

# Execute order book fetching
fetch_order_books()
```

#### Fetch Market Prices from CLOB

```python
# Cell 47: Fetch Market Prices from CLOB
def fetch_clob_prices():
    """Fetch buy/sell prices and midpoints for all tokens"""

    clob_url = "https://clob.polymarket.com"
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get all token IDs
    cursor.execute("""
        SELECT DISTINCT
            m.clob_token_ids,
            m.condition_id,
            m.question,
            m.outcomes
        FROM markets_enhanced m
        WHERE m.active = 1
        AND m.clob_token_ids IS NOT NULL
        AND m.clob_token_ids != '[]'
        ORDER BY m.volume DESC
        LIMIT 200
    """)

    markets = cursor.fetchall()

    print(f"💹 Fetching prices for {len(markets)} markets...")
    print("="*60)

    # First, get all prices at once
    print("  📊 Fetching bulk prices...")
    try:
        url = f"{clob_url}/prices"
        response = requests.get(url)

        if response.status_code == 200:
            all_prices = response.json()

            # Store bulk prices
            for token_id, prices in all_prices.items():
                buy_price = float(prices.get('BUY', 0)) if prices.get('BUY') else None
                sell_price = float(prices.get('SELL', 0)) if prices.get('SELL') else None

                if buy_price and sell_price:
                    spread = sell_price - buy_price
                    mid_price = (buy_price + sell_price) / 2

                    cursor.execute("""
                    INSERT OR REPLACE INTO clob_prices
                    (token_id, buy_price, sell_price, mid_price, spread)
                    VALUES (?, ?, ?, ?, ?)
                    """, (token_id, buy_price, sell_price, mid_price, spread))

            print(f"     Bulk prices stored: {len(all_prices)} tokens")
    except Exception as e:
        print(f"     ⚠️ Bulk price fetch failed: {e}")

    # Now fetch individual prices with midpoints for top markets
    print("\n  📊 Fetching individual prices and midpoints...")

    total_prices = 0
    price_data = []

    for idx, (token_ids_str, condition_id, question, outcomes_str) in enumerate(markets[:100], 1):
        try:
            token_ids = json.loads(token_ids_str) if token_ids_str else []
            outcomes = json.loads(outcomes_str) if outcomes_str else []

            for i, token_id in enumerate(token_ids[:2]):  # Get first 2 tokens
                try:
                    # Get buy price
                    url = f"{clob_url}/price"
                    buy_response = requests.get(url, params={"token_id": token_id, "side": "BUY"})

                    # Get sell price
                    sell_response = requests.get(url, params={"token_id": token_id, "side": "SELL"})

                    # Get midpoint
                    mid_url = f"{clob_url}/midpoint"
                    mid_response = requests.get(mid_url, params={"token_id": token_id})

                    if buy_response.status_code == 200 and sell_response.status_code == 200:
                        buy_data = buy_response.json()
                        sell_data = sell_response.json()

                        buy_price = float(buy_data.get('price', 0)) if buy_data.get('price') else None
                        sell_price = float(sell_data.get('price', 0)) if sell_data.get('price') else None

                        mid_price = None
                        if mid_response.status_code == 200:
                            mid_data = mid_response.json()
                            mid_price = float(mid_data.get('mid', 0)) if mid_data.get('mid') else None

                        if buy_price and sell_price:
                            spread = sell_price - buy_price

                            # Store prices
                            cursor.execute("""
                            INSERT OR REPLACE INTO clob_prices
                            (token_id, buy_price, sell_price, mid_price, spread)
                            VALUES (?, ?, ?, ?, ?)
                            """, (token_id, buy_price, sell_price, mid_price or (buy_price + sell_price) / 2, spread))

                            total_prices += 1

                            outcome = outcomes[i] if i < len(outcomes) else f"Option {i+1}"

                            price_data.append({
                                'question': question,
                                'outcome': outcome,
                                'token_id': token_id,
                                'buy': buy_price,
                                'sell': sell_price,
                                'mid': mid_price or (buy_price + sell_price) / 2,
                                'spread': spread
                            })

                            if idx == 1 and i == 0:
                                print(f"  ✓ Example: {question[:50]}...")
                                print(f"     Outcome: {outcome}")
                                print(f"     Buy: {buy_price:.4f} | Sell: {sell_price:.4f}")
                                print(f"     Mid: {mid_price or (buy_price + sell_price) / 2:.4f}")
                                print(f"     Spread: {spread:.4f} ({spread*100:.2f}%)")

                except Exception as e:
                    # Silently skip token errors
                    pass

        except Exception as e:
            # Silently skip market errors
            pass

        # Progress update
        if idx % 20 == 0 or idx == min(100, len(markets)):
            print(f"  📊 Progress: {idx}/{min(100, len(markets))} markets processed")
            print(f"     Prices fetched: {total_prices}")
            conn.commit()

    # Sort by tightest spreads
    price_data.sort(key=lambda x: x['spread'])

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(DISTINCT token_id) as tokens_with_prices,
            AVG(buy_price) as avg_buy,
            AVG(sell_price) as avg_sell,
            AVG(mid_price) as avg_mid,
            AVG(spread) as avg_spread,
            MIN(spread) as min_spread,
            MAX(spread) as max_spread
        FROM clob_prices
        WHERE timestamp = (SELECT MAX(timestamp) FROM clob_prices)
    """)

    stats = cursor.fetchone()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Price fetch complete!")
    print(f"   Markets processed: {min(100, len(markets))}")
    print(f"   Individual prices fetched: {total_prices}")

    if stats:
        tokens, avg_buy, avg_sell, avg_mid, avg_spread, min_spread, max_spread = stats
        print(f"\n📊 Price Statistics:")
        print(f"   Tokens with prices: {tokens}")
        print(f"   Average buy price: {avg_buy or 0:.4f}")
        print(f"   Average sell price: {avg_sell or 0:.4f}")
        print(f"   Average mid price: {avg_mid or 0:.4f}")
        print(f"   Average spread: {avg_spread or 0:.4f}")
        print(f"   Tightest spread: {min_spread or 0:.4f}")
        print(f"   Widest spread: {max_spread or 0:.4f}")

    print(f"\n💹 Markets with Tightest Spreads:")
    for i, data in enumerate(price_data[:10], 1):
        print(f"   {i}. {data['question'][:40]}... - {data['outcome']}")
        print(f"      Buy: {data['buy']:.4f} | Sell: {data['sell']:.4f} | Mid: {data['mid']:.4f}")
        print(f"      Spread: {data['spread']:.4f} ({data['spread']*100:.2f}%)")

# Execute price fetching
fetch_clob_prices()
```

#### Fetch Price History from CLOB

```python
# Cell 48: Fetch Price History from CLOB
def fetch_price_history():
    """Fetch historical price data for top markets"""

    clob_url = "https://clob.polymarket.com"
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get top markets
    cursor.execute("""
        SELECT DISTINCT
            m.clob_token_ids,
            m.question,
            m.outcomes
        FROM markets_enhanced m
        WHERE m.active = 1
        AND m.clob_token_ids IS NOT NULL
        AND m.clob_token_ids != '[]'
        AND m.volume > 100000
        ORDER BY m.volume DESC
        LIMIT 50
    """)

    markets = cursor.fetchall()

    print(f"📈 Fetching price history for {len(markets)} top markets...")
    print("="*60)

    # Time periods to fetch
    current_ts = int(time.time())
    periods = {
        "1 day": {"interval": "1d", "fidelity": 60},  # 60 min intervals
        "1 week": {"interval": "1w", "fidelity": 240},  # 4 hour intervals
        "1 month": {"interval": "1m", "fidelity": 1440}  # Daily intervals
    }

    total_history_points = 0
    volatile_markets = []

    for idx, (token_ids_str, question, outcomes_str) in enumerate(markets, 1):
        try:
            token_ids = json.loads(token_ids_str) if token_ids_str else []
            outcomes = json.loads(outcomes_str) if outcomes_str else []

            for i, token_id in enumerate(token_ids[:2]):  # Get first 2 tokens
                try:
                    # Fetch 1 week history
                    url = f"{clob_url}/prices-history"
                    params = {
                        "market": token_id,
                        "interval": "1w",
                        "fidelity": 240  # 4 hour intervals
                    }

                    response = requests.get(url, params=params)

                    if response.status_code == 200:
                        data = response.json()
                        history = data.get('history', [])

                        if history:
                            # Store history points
                            for point in history:
                                cursor.execute("""
                                INSERT OR REPLACE INTO price_history
                                (token_id, timestamp, price)
                                VALUES (?, ?, ?)
                                """, (token_id, point['t'], point['p']))

                            total_history_points += len(history)

                            # Calculate volatility
                            prices = [p['p'] for p in history]
                            if len(prices) > 1:
                                min_price = min(prices)
                                max_price = max(prices)
                                avg_price = sum(prices) / len(prices)
                                volatility = (max_price - min_price) / avg_price if avg_price > 0 else 0

                                outcome = outcomes[i] if i < len(outcomes) else f"Option {i+1}"

                                if volatility > 0.2:  # >20% volatility
                                    volatile_markets.append({
                                        'question': question,
                                        'outcome': outcome,
                                        'volatility': volatility,
                                        'min': min_price,
                                        'max': max_price,
                                        'current': prices[-1],
                                        'change': prices[-1] - prices[0]
                                    })

                                if idx == 1 and i == 0:
                                    print(f"  ✓ Example: {question[:50]}...")
                                    print(f"     Outcome: {outcome}")
                                    print(f"     History points: {len(history)}")
                                    print(f"     Price range: {min_price:.4f} - {max_price:.4f}")
                                    print(f"     Current: {prices[-1]:.4f}")
                                    print(f"     Change: {(prices[-1] - prices[0]):.4f}")
                                    print(f"     Volatility: {volatility*100:.1f}%")

                except Exception as e:
                    # Silently skip token errors
                    pass

        except Exception as e:
            # Silently skip market errors
            pass

        # Progress update
        if idx % 10 == 0 or idx == len(markets):
            print(f"  📊 Progress: {idx}/{len(markets)} markets processed")
            print(f"     History points stored: {total_history_points}")
            conn.commit()

    # Sort volatile markets
    volatile_markets.sort(key=lambda x: x['volatility'], reverse=True)

    # Get statistics
    cursor.execute("""
        SELECT
            COUNT(DISTINCT token_id) as tokens_with_history,
            COUNT(*) as total_points,
            MIN(timestamp) as earliest_time,
            MAX(timestamp) as latest_time,
            AVG(price) as avg_price,
            MIN(price) as min_price,
            MAX(price) as max_price
        FROM price_history
    """)

    stats = cursor.fetchone()

    # Get biggest movers
    cursor.execute("""
        WITH price_changes AS (
            SELECT
                token_id,
                MIN(price) as min_price,
                MAX(price) as max_price,
                FIRST_VALUE(price) OVER (PARTITION BY token_id ORDER BY timestamp) as first_price,
                LAST_VALUE(price) OVER (PARTITION BY token_id ORDER BY timestamp
                    ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) as last_price
            FROM price_history
            GROUP BY token_id
        )
        SELECT
            token_id,
            min_price,
            max_price,
            first_price,
            last_price,
            (last_price - first_price) as absolute_change,
            (last_price - first_price) / NULLIF(first_price, 0) * 100 as percent_change
        FROM price_changes
        WHERE ABS(last_price - first_price) > 0.1
        ORDER BY ABS(percent_change) DESC
        LIMIT 10
    """)

    biggest_movers = cursor.fetchall()

    conn.commit()
    conn.close()

    print("="*60)
    print(f"✅ Price history fetch complete!")
    print(f"   Markets processed: {len(markets)}")
    print(f"   History points stored: {total_history_points}")

    if stats:
        tokens, points, earliest, latest, avg, min_p, max_p = stats
        print(f"\n📊 Historical Data Statistics:")
        print(f"   Tokens with history: {tokens}")
        print(f"   Total data points: {points}")
        if earliest and latest:
            time_range = (latest - earliest) / 86400  # Convert to days
            print(f"   Time range: {time_range:.1f} days")
        print(f"   Price range: {min_p or 0:.4f} - {max_p or 0:.4f}")
        print(f"   Average price: {avg or 0:.4f}")

    print(f"\n📈 Most Volatile Markets (>20% volatility):")
    for i, market in enumerate(volatile_markets[:10], 1):
        print(f"   {i}. {market['question'][:40]}... - {market['outcome']}")
        print(f"      Volatility: {market['volatility']*100:.1f}%")
        print(f"      Range: {market['min']:.4f} - {market['max']:.4f}")
        print(f"      Current: {market['current']:.4f} | Change: {market['change']:+.4f}")

    print(f"\n🚀 Biggest Price Movers:")
    for i, (token_id, min_p, max_p, first, last, abs_change, pct_change) in enumerate(biggest_movers, 1):
        symbol = "📈" if abs_change > 0 else "📉"
        print(f"   {i}. {symbol} Token: {token_id[:20]}...")
        print(f"      Change: {abs_change:+.4f} ({pct_change or 0:+.1f}%)")
        print(f"      From {first:.4f} → {last:.4f}")

# Execute price history fetching
fetch_price_history()
```

## **Spreads**

#### Fetch Bid-Ask Spreads from CLOB

```python
# Cell 49: Fetch Bid-Ask Spreads from CLOB
def fetch_spreads():
    """Fetch bid-ask spreads for all active markets"""

    clob_url = "https://clob.polymarket.com"
    db_path = os.getenv("DB_PATH")

    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    print(f"📊 Fetching bid-ask spreads from CLOB...")
    print("="*60)

    try:
        # Fetch all spreads at once
        url = f"{clob_url}/spreads"
        headers = {"Content-Type": "application/json"}

        response = requests.post(url, headers=headers)

        if response.status_code == 200:
            spreads_data = response.json()

            print(f"  ✓ Received spreads for {len(spreads_data)} tokens")

            # Process and store spreads
            tight_spreads = []
            wide_spreads = []
            total_stored = 0

            for token_id, spread_str in spreads_data.items():
                try:
                    spread = float(spread_str)

                    # Get current prices for this token to calculate bid/ask
                    cursor.execute("""
                        SELECT buy_price, sell_price
                        FROM clob_prices
                        WHERE token_id = ?
                        ORDER BY timestamp DESC
                        LIMIT 1
                    """, (token_id,))

                    result = cursor.fetchone()

                    if result:
                        buy_price, sell_price = result
                    else:
                        # Estimate from spread and midpoint
                        mid_price = 0.5  # Default assumption
                        buy_price = mid_price - spread/2
                        sell_price = mid_price + spread/2

                    # Store spread
                    cursor.execute("""
                    INSERT INTO market_spreads
                    (token_id, spread, bid_price, ask_price)
                    VALUES (?, ?, ?, ?)
                    """, (token_id, spread, buy_price or 0, sell_price or 1))

                    total_stored += 1

                    # Track tight and wide spreads
                    if spread < 0.01:  # <1% spread
                        tight_spreads.append({
                            'token_id': token_id,
                            'spread': spread,
                            'bid': buy_price,
                            'ask': sell_price
                        })
                    elif spread > 0.1:  # >10% spread
                        wide_spreads.append({
                            'token_id': token_id,
                            'spread': spread,
                            'bid': buy_price,
                            'ask': sell_price
                        })

                except Exception as e:
                    # Skip invalid spread values
                    pass

            print(f"  ✓ Stored {total_stored} spreads in database")

            # Sort spreads
            tight_spreads.sort(key=lambda x: x['spread'])
            wide_spreads.sort(key=lambda x: x['spread'], reverse=True)

            # Get market info for tokens with best spreads
            if tight_spreads:
                token_list = [s['token_id'] for s in tight_spreads[:10]]
                token_str = "','".join(token_list)

                cursor.execute(f"""
                    SELECT DISTINCT m.question, m.clob_token_ids
                    FROM markets_enhanced m
                    WHERE m.clob_token_ids LIKE '%' || ? || '%'
                    LIMIT 10
                """, (token_list[0],))

                market_info = cursor.fetchall()

            # Get statistics
            cursor.execute("""
                SELECT
                    COUNT(*) as total_spreads,
                    AVG(spread) as avg_spread,
                    MIN(spread) as min_spread,
                    MAX(spread) as max_spread,
                    COUNT(CASE WHEN spread < 0.01 THEN 1 END) as tight_count,
                    COUNT(CASE WHEN spread > 0.1 THEN 1 END) as wide_count,
                    COUNT(CASE WHEN spread > 0.5 THEN 1 END) as very_wide_count
                FROM market_spreads
                WHERE timestamp = (SELECT MAX(timestamp) FROM market_spreads)
            """)

            stats = cursor.fetchone()

            # Get spread distribution
            cursor.execute("""
                SELECT
                    COUNT(CASE WHEN spread < 0.005 THEN 1 END) as ultra_tight,
                    COUNT(CASE WHEN spread >= 0.005 AND spread < 0.01 THEN 1 END) as very_tight,
                    COUNT(CASE WHEN spread >= 0.01 AND spread < 0.05 THEN 1 END) as tight,
                    COUNT(CASE WHEN spread >= 0.05 AND spread < 0.1 THEN 1 END) as moderate,
                    COUNT(CASE WHEN spread >= 0.1 AND spread < 0.2 THEN 1 END) as wide,
                    COUNT(CASE WHEN spread >= 0.2 THEN 1 END) as very_wide
                FROM market_spreads
                WHERE timestamp = (SELECT MAX(timestamp) FROM market_spreads)
            """)

            distribution = cursor.fetchone()

            conn.commit()
            conn.close()

            print("="*60)
            print(f"✅ Spread fetch complete!")
            print(f"   Total spreads fetched: {len(spreads_data)}")
            print(f"   Spreads stored: {total_stored}")

            if stats:
                total, avg, min_s, max_s, tight, wide, very_wide = stats
                print(f"\n📊 Spread Statistics:")
                print(f"   Average spread: {avg or 0:.4f} ({(avg or 0)*100:.2f}%)")
                print(f"   Tightest spread: {min_s or 0:.4f} ({(min_s or 0)*100:.2f}%)")
                print(f"   Widest spread: {max_s or 0:.4f} ({(max_s or 0)*100:.2f}%)")
                print(f"   Tight spreads (<1%): {tight}")
                print(f"   Wide spreads (>10%): {wide}")
                print(f"   Very wide (>50%): {very_wide}")

            if distribution:
                ultra, very_t, tight, moderate, wide, very_w = distribution
                print(f"\n📈 Spread Distribution:")
                print(f"   Ultra-tight (<0.5%): {ultra} tokens")
                print(f"   Very tight (0.5-1%): {very_t} tokens")
                print(f"   Tight (1-5%): {tight} tokens")
                print(f"   Moderate (5-10%): {moderate} tokens")
                print(f"   Wide (10-20%): {wide} tokens")
                print(f"   Very wide (>20%): {very_w} tokens")

            print(f"\n💹 Markets with Tightest Spreads:")
            for i, spread_data in enumerate(tight_spreads[:10], 1):
                print(f"   {i}. Token: {spread_data['token_id'][:20]}...")
                print(f"      Spread: {spread_data['spread']:.5f} ({spread_data['spread']*100:.3f}%)")
                if spread_data.get('bid'):
                    print(f"      Bid: {spread_data['bid']:.4f} | Ask: {spread_data.get('ask', 0):.4f}")

            print(f"\n⚠️  Markets with Widest Spreads (low liquidity):")
            for i, spread_data in enumerate(wide_spreads[:5], 1):
                print(f"   {i}. Token: {spread_data['token_id'][:20]}...")
                print(f"      Spread: {spread_data['spread']:.4f} ({spread_data['spread']*100:.1f}%)")

        else:
            print(f"  ❌ Failed to fetch spreads: {response.status_code}")
            conn.close()

    except Exception as e:
        print(f"  ❌ Error fetching spreads: {e}")
        conn.close()

# Execute spread fetching
fetch_spreads()
```
